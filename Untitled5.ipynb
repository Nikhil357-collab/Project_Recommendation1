```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import IsolationForest

```

# New section

```python
#Step 2. Load data
df = pd.read_csv("/content/fraud_detection.csv")
df.head()
#Step 3. Basic checks
df.shape
print(df.columns)
#df['TransactionID'].value_counts() # This line caused the error. After inspecting df.columns, replace 'Class' with the correct column name.
```

```python
# 3. Missing value percentage
missing_pct = df.isnull().mean() * 100
print("\nMissing Value Percentage (%):")
print(missing_pct)
```

```python
cat_cols = [
    'TransactionType', 'Location', 'DeviceID',
    'MerchantID', 'Channel', 'CustomerOccupation'
]

le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
print(df.columns)
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['PreviousTransactionDate'] = pd.to_datetime(df['PreviousTransactionDate'])

df['DaysSinceLastTransaction'] = (
    df['TransactionDate'] - df['PreviousTransactionDate']
).dt.days
```

```python
df.drop(['TransactionDate', 'PreviousTransactionDate'], axis=1, inplace=True)
```

```python
info = df.info()
head = df.head()
```

```python
scaler = StandardScaler()

# Identify columns that are not suitable for numerical scaling
# These include identifier columns (TransactionID, AccountID, IP Address) and the boolean flag 'is_outlier'
columns_to_exclude = ['TransactionID', 'AccountID', 'IP Address', 'is_outlier']

# Create a subset of the DataFrame containing only the columns to be scaled
df_to_scale = df.drop(columns=columns_to_exclude)

df_scaled = scaler.fit_transform(df_to_scale)
```

```python
iso = IsolationForest(
    n_estimators=200,
    contamination=0.01,   # assume 1% fraud
    random_state=42
)

df['Anomaly'] = iso.fit_predict(df_scaled)
```

```python
df['FraudFlag'] = df['Anomaly'].map({1: 0, -1: 1})
df['FraudFlag'].value_counts()
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(contamination=0.01)
df['Fraud_LOF'] = lof.fit_predict(df_scaled)
```

```python
iso = IsolationForest(
    n_estimators=200,
    contamination=0.01,   # assume 1% fraud
    random_state=42
)

df['Anomaly'] = iso.fit_predict(df_scaled)
```

```python
df['FraudFlag'] = df['Anomaly'].map({1: 0, -1: 1})
df['FraudFlag'].value_counts()
```

```python
sns.boxplot(x='FraudFlag', y='TransactionAmount', data=df)
plt.show()
```

```python
# 2. Descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe(include="all"))
```

```python
# 4. Plot distributions (numeric columns only)
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure()
    plt.hist(df[col].dropna())
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# 5. Detect outliers using IQR
outlier_flags = pd.DataFrame(index=df.index)

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outlier_flags[col + "_outlier"] = ((df[col] < lower) | (df[col] > upper))
```

```python
# 6. Create overall outlier flag column
df["is_outlier"] = outlier_flags.any(axis=1)

print("\nOutlier count:")
print(df["is_outlier"].value_counts())
```
